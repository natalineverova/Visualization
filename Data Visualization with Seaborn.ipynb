```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Read in the DataFrame
df = pd.read_csv('grant_file.csv')
df2 = pd.read_csv('insurance_premiums.csv')
```

```python
# Display pandas histogram
df['Award_Amount'].plot.hist()
plt.show()

# Clear out the pandas histogram
plt.clf()
```

```python
# Display a Seaborn distplot
sns.distplot(df['Award_Amount'])
plt.show()

# Clear the distplot
plt.clf();
```

```python
# Create a distplot
sns.distplot(df['Award_Amount'],
             kde=False,
             bins=20)

# Display the plot
plt.show();
```

```python
# Create a distplot of the Award Amount
sns.distplot(df['Award_Amount'],
             hist=False,
             rug=True,
             kde_kws={'shade':True});

# Plot the results
plt.show();
```

```python
# Create a regression plot of premiums vs. insurance_losses
sns.regplot(x="insurance_losses", y="premiums", data=df2)


# Display the plot
plt.show()
```

```python
# Create an lmplot of premiums vs. insurance_losses
sns.lmplot(x="insurance_losses", y="premiums", data=df2)



# Display the second plot
plt.show()
```

```python
# Create a regression plot using hue
sns.lmplot(data=df2,
           x="insurance_losses",
           y="premiums",
           hue ="Region")

# Show the results
plt.show()
```

```python
# Create a regression plot with multiple rows
sns.lmplot(data=df2,
           x="insurance_losses",
           y="premiums",
           col="Region") #row - строками

# Show the plot
plt.show()
```

```python
g = sns.PairGrid(df2, vars=["fatal_collisions", "premiums"])
g2 = g.map(plt.scatter)

plt.show()
plt.clf()
```

```python
# Create the same PairGrid but map a histogram on the diag
g = sns.PairGrid(df2, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)

plt.show()
plt.clf()
```

```python
# Create a pairwise plot of the variables using a scatter plot
sns.pairplot(data=df2,
        vars=["fatal_collisions", "premiums"],
        kind='scatter')

plt.show();
plt.clf()
```

```python
# Plot the same data but use a different color palette and color code by Region
sns.pairplot(data=df2,
        vars=["fatal_collisions", "premiums"],
        kind='scatter',
        hue='Region',
        palette='RdBu',
        diag_kws={'alpha':.5})

plt.show()
plt.clf()
```
